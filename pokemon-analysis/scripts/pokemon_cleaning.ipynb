```python
#Pandas for cleaning, IPython for displaying nicer tables

import pandas as pd
from IPython.display import display
```

```python
#Getting an understanding of the dataset

pokedex_data = pd.read_csv("pokedex.csv")
pokedex_data.info() #get info on the types 
pokedex_data.columns
pokedex_data.shape
```

```python
# Rename "Unamed: 0" to "id"
pokedex_data = pokedex_data.rename(columns={"Unnamed: 0": "id"})
pokedex_data.columns
```

```python
#quick backup and tested on mysql
pokedex_data.to_csv('cleaned_pokedex.csv', index=False)
print("Cleaned data saved as 'Cleaned_pokedex.csv'")
```

```python
#checking why SQL import failed - some columns had data unrecognisable e.g. japanese_name
display(pokedex_data.head(5))
```

```python
#dropping unrecognisable columns and columns I don't intend on using

pokedex_data.drop(columns=['german_name', 'japanese_name', 'egg_type_number', 'egg_type_1',
       'egg_type_2', 'percentage_male', 'egg_cycles'], inplace=True)
pokedex_data.columns
```

```python
pokedex_data.to_csv('cleaned_pokedex.csv', index=False)
print("Cleaned data saved as 'Cleaned_pokedex.csv'")
```

```python
#cleaning up the species column

pokedex_data['species'] = pokedex_data['species'].str.replace('Pokémon', '', case=True).str.strip()
display(pokedex_data.head(5))
```

```python
#cleaning up the ability columns with none if there is null

print(pokedex_data.isnull().sum())
pokedex_data["type_2"].fillna("None", inplace=True)
pokedex_data["ability_1"].fillna("None", inplace=True)
pokedex_data["ability_2"].fillna("None", inplace=True)
pokedex_data["ability_hidden"].fillna("None", inplace=True)
print(pokedex_data.isnull().sum())
```
